```python
import matplotlib.pyplot as plt
import numpy as np

# Rastgele noktalar oluştur
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)

# Rastgele boyutlar ve renkler
sizes = np.random.randint(100, 800, size=100)
colors = np.random.rand(100)

# Dikenli harita plotlaması
plt.figure(figsize=(8, 8))
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.8, marker='^')
plt.colorbar()

plt.title("Dikenli Harita Plotlaması")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

# Veri noktaları
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Hareketli grafik için rastgele değerler
np.random.seed(42)
offsets = np.random.normal(0, 0.5, size=(100,))

# Grafik nesnesi oluşturma
fig, ax = plt.subplots()

# Boş bir çizgi oluştur
line, = ax.plot([], [])

# Veri noktalarını güncelleyen fonksiyon
def update(frame):
    line.set_data(x, y + offsets[frame])
    ax.set_xlim(0, 2 * np.pi)
    ax.set_ylim(-2, 2)
    ax.set_title("Hareketli Grafik")
    ax.set_xlabel("X Ekseni")
    ax.set_ylabel("Y Ekseni")

# Animasyonu güncelleyen fonksiyon
def animate(frame):
    update(frame)
    return line,

# Animasyonu oluşturma
ani = animation.FuncAnimation(fig, animate, frames=len(offsets), interval=200, blit=True)

# Animasyonu gösterme
plt.show()



```
